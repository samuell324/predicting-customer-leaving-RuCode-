### Бинарное кодирование колонок "скидки" и "жалобы"

```python
train_data['скидки'] = train_data['скидки'].apply(lambda x: 1 if x == 'Yes' else 0)
train_data['жалобы'] = train_data['жалобы'].apply(lambda x: 1 if x == 'Yes' else 0)
```

### Работа с пропусками

```python
#Удаление строк для которых неизвестен пол
train_data = train_data[train_data['пол']!='Unknown']
train_data.head(10)
```

```python
#бинарное кодирование признака "пол"
train_data['пол'] = train_data['пол'].apply(lambda x: 1 if x =='M' else 0)
```

```python
#Если не указан тип региона проживания -> устанавливаем в значение Town
train_data['тип_региона_проживания'] = train_data['тип_региона_проживания'].fillna('Town')
```

```python
train_data['баллы'].hist()
```

```python
#заполнение пропусков признака "баллы"
mean = train_data['баллы'].mean()
train_data['баллы'] = train_data['баллы'].fillna(mean)
```

```python
#Реферальная ссылка. Если о человеке нет информации -> нет реферальной ссылки
mask = train_data['реферальная_ссылка']=='?'
train_data.loc[mask, 'реферальная_ссылка'] = 'No'
```

```python
#бинарное кодирование признака "реферальная_ссылка"
train_data['реферальная_ссылка'] = train_data['реферальная_ссылка'].apply(lambda x: 1 if x =='Yes' else 0)
```

```python
train_data.head(10)
```

### Перевод категориальных признаков в числовые

```python
#Что осталось?
categorical_columns = [x for x in train_data.columns if train_data[x].dtype =='object']
categorical_columns
```

```python
train_data = pd.get_dummies(train_data, categorical_columns)
train_data.head()
```

```python
#Делим данные на Х и У
X = train_data.drop('риск_ухода', axis=1)
y = train_data['риск_ухода']
```

## Построение моделей

```python
from sklearn import ensemble, linear_model
from sklearn.model_selection import cross_val_score
from sklearn.metrics import mean_squared_error
```

```python
linreg = linear_model.LinearRegression()
rf = ensemble.RandomForestRegressor()
gb = ensemble.GradientBoostingRegressor()
```

```python
linreg.fit(X,y)
rf.fit(X,y)
gb.fit(X,y)
```

```python
linreg.score(X,y)
```

```python
rf.score(X,y)
```

```python
gb.score(X,y)
```

```python
gb = ensemble.GradientBoostingRegressor(n_estimators =500, max_depth=5, min_samples_split=10)
gb.fit(X,y)
```

```python
gb.score(X,y)
```

```python
from catboost import CatBoostRegressor
model = CatBoostRegressor(iterations=100,
                          learning_rate=1,
                          depth=5,
                         loss_function = 'RMSE')
```
